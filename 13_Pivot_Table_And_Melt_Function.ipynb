# Melt

```python
import pandas as pd
```

```python
var=pd.DataFrame({"Days":[1,2,3,4,5,6],
                  "Maths":[23,21,19,12,20,14],
                  "English":[13,12,23,12,25,21],
                  "Physics":[11,21,12,13,14,15]})
var
```

```python
var.melt()
```

```python
var.melt(id_vars="Days",var_name="Subjects",value_name="Marks")
```

# Pivot Table 

```python
var1=pd.DataFrame({"Days":[1,2,3,4,5,6],
                   "Student Name":["Ali","Ahmed","Munir","Qasim","Asad","Saad"],
                  "Maths":[23,21,19,12,20,14],
                  "English":[13,12,23,12,25,21],
                  "Physics":[11,21,12,13,14,15]})
var1
```

```python
var1.pivot(index="Days",columns="Student Name")
```

```python
var1.pivot(index="Days",columns="Student Name",values="Maths")
```

```python
var1=pd.DataFrame({"Days":[1,2,3,4,5,6],
                   "Student Name":["Ali","Ahmed","Munir","Qasim","Asad","Saad"],
                  "Maths":[23,21,19,12,20,14],
                  "English":[13,12,23,12,25,21],
                  "Physics":[11,21,12,13,14,15]})
var1
```

```python
var1.pivot_table(index="Student Name",columns="Days",aggfunc="mean")
```

```python
var1.pivot_table(index="Student Name",columns="Days",aggfunc="sum")
```

```python
var1.pivot_table(index="Student Name",columns="Days",aggfunc="sum",margins=True)
```
