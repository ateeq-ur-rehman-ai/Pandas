# Replace 

```python
import pandas as pd
```

```python
var =pd.read_csv("Test_File_2.csv")
```

```python
var
```

```python
var.replace(to_replace=88,value=1000)
```

```python
var.replace(to_replace="Rahul",value="PANDAS")
```

```python
var.replace([101,102,103,104,105,106,107,108,109,110],"*")
```

```python
var.replace("[A-Za-z]","Hello",regex=True)
```

```python
var.replace({"City" : "[A-z]"},100,regex=True)
```

```python
var.replace("Mumbai",method="ffill")
```

```python
var.replace("Mumbai",method="ffill",limit=1,inplace=True)
```

```python
var
```

# Interpolate

```python
import pandas as pd
```

```python
var1=pd.read_csv("Test_File_4.csv")
var1
```

```python
var1.interpolate()
# it only work with numbers or digits
```

```python
var1.interpolate(method="linear")
```

```python
var1.interpolate(method="linear",axix=0)
```

```python
var1.interpolate(method="linear",axis=1)
```

```python
var1.interpolate(limit_direction="forward",limit=1)
```

```python
var1.interpolate(limit_direction="backward",limit=1)
```

```python
var1.interpolate(limit_direction="both",limit=1)
```

```python
var1.interpolate(limit_direction="both",limit=2,inplace=True)
```
