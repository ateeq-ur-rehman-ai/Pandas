# Merge

```python
import pandas as pd 
```

```python
var1=pd.DataFrame({"A":[23,1,4,31],"B":[24,68,57,43]})
var2=pd.DataFrame({"A":[23,1,4,31],"C":[23,75,85,36]})
pd.merge(var1,var2,on="A")
```

```python
var1=pd.DataFrame({"A":[23,1,4,31],"B":[24,68,57,43]})
var2=pd.DataFrame({"A":[23,1,4,11],"C":[23,75,85,36]})
pd.merge(var1,var2,on="A")
```

```python
var1=pd.DataFrame({"A":[23,1,4,31],"B":[24,68,57,43]})
var2=pd.DataFrame({"A":[23,1,4,11],"C":[23,75,85,36]})
pd.merge(var1,var2)
```

```python
var1=pd.DataFrame({"A":[23,1,4,31],"B":[24,68,57,43]})
var2=pd.DataFrame({"A":[23,1,4,11],"C":[23,75,85,36]})
pd.merge(var1,var2,how="left")
```

```python
var1=pd.DataFrame({"A":[23,1,4,31],"B":[24,68,57,43]})
var2=pd.DataFrame({"A":[23,1,4,11],"C":[23,75,85,36]})
pd.merge(var1,var2,how="right")

# by defualt inner
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[24,68,57,43]})
var2=pd.DataFrame({"A":[1,2,3,5],"C":[23,75,85,36]})
pd.merge(var1,var2,how="outer", indicator=True)
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[24,68,57,43]})
var2=pd.DataFrame({"A":[1,2,3,5],"B":[23,75,85,36]})
pd.merge(var1,var2,left_index=True,right_index=True)
```

```python
var1=pd.DataFrame({"A":[1,2,3,4],"B":[24,68,57,43]})
var2=pd.DataFrame({"A":[1,2,3,5],"B":[23,75,85,36]})
pd.merge(var1,var2,left_index=True,right_index=True,suffixes=("Name","ID"))
```

# Concat

```python
sr1=pd.Series([1,2,3,4])
sr2=pd.Series([3,2,51,42])

pd.concat([sr1,sr2])
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[24,68,57,43]})
d2=pd.DataFrame({"A":[1,2,3,5],"B":[23,75,85,36]})
pd.concat([d1,d2])
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[24,68,57,43]})
d2=pd.DataFrame({"A":[1,2,3,5],"B":[23,75,85,36]})
pd.concat([d1,d2],axis=1)
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[24,68,57,43]})
d2=pd.DataFrame({"A":[1,2,3,5],"B":[23,75,85,36]})
pd.concat([d1,d2],axis=0)
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[24,68,57,43]})
d2=pd.DataFrame({"A":[1,2,],"B":[23,75]})
pd.concat([d1,d2],axis=1,join="inner")
```

```python
d1=pd.DataFrame({"A":[1,2,3,4],"B":[24,68,57,43]})
d2=pd.DataFrame({"A":[1,2,3,5],"B":[23,75,85,36]})
pd.concat([d1,d2],axis=1,keys=["D1","D2"])
```
