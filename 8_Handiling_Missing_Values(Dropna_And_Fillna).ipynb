# Dropna

```python
import pandas as pd
```

```python
csv=pd.read_csv("Test_File_3.csv")
csv
```

```python
csv.dropna()
```

```python
csv.dropna(axis=1)
```

```python
csv.dropna(how="any")
```

```python
csv.dropna(how="all")
```

```python
csv.dropna(inplace=True)
csv
```

```python
csv.dropna(thresh=1)
```

 # Fillna

```python
csv1=pd.read_csv("Test_File_3.csv")
csv1
```
